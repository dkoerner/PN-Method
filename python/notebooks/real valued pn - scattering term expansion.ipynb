&= \sigma_s(\vec{x})(L\circ \rho_{R(\omega)}(f_p))(\omega) \\
\end{align}

$$
\begin{align}
(L\circ \rho_{R(\omega)}(f_p)) &= \int_{\Omega'}{L(\vec{x}, \omega')\rho_{R(\omega)}(f_p)(\omega')\mathbf{d}\omega'} \\
&= \langle L,  \rho_{R(\omega)}(f_p)\rangle
\end{align}
