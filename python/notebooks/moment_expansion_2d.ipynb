<h2><center>Moment Expansion in 2D</center></h2>

The moment expansion follows closely its 3d counterpart, which is discussed in the notebook "moment_expansion_3d". First we will derive the 2d counterpart of the spherical harmonics expansion on a circular function. Then we will derive the components of the two dimensional moment tensors by expression the expansion in cartesian coordinates.

The two dimensional equivalent of the spherical harmonics expansion of a spherical function is the fourier expansion of a one dimensional periodic function with period $2\pi$. We will refer to it as a circular function. Such a function can be expressed in terms of its fourier bases:

$$
\begin{align}
f(t) = a_0 &+ a_1\operatorname{cos}(t) + a_2\operatorname{cos}(2t) + ... + a_n\operatorname{cos}(nt)\\
           &+ b_1\operatorname{sin}(t) + b_2\operatorname{sin}(2t) + ... + b_n\operatorname{sin}(nt)
\end{align}
$$

The function $f$ can be projected into its fourier coefficients by integration:
$$
a_0=\frac{1}{2\pi}\int_0^{2\pi}{f(t)\mathrm{dt}}
$$
$$
a_n=\frac{1}{\pi}\int_0^{2\pi}{f(t)\operatorname{cos}(nt)\mathrm{dt}}
$$
$$
b_n=\frac{1}{\pi}\int_0^{2\pi}{f(t)\operatorname{sin}(nt)\mathrm{dt}}
$$

----------------------------------

Like in the three dimensional case, we can also expand $f(t)$ in terms of its moments. For a two dimensional unit direction vector $\vec{n}$ we have:
$$
f(\vec{n}) = \mu_0[f] + \mu_1[f]\vec{n}_i + \mu_2[f]\vec{n}_{i}\vec{n}_{j}
$$

Here $\mu_n[f]$ is the n-th moment of $f$, a two-dimensional rank-$n$ tensor. The operation in each term on the RHS are tensor contractions (see survey, chapter 3.1 for details).

In order to find the tensor components for $\mu_n$ we follow the same procedure as in the three-dimensional case. We first give the fourier bases which is related to moment $n$:

$$
\operatorname{fourierbasis}_0(t) = a_0
\operatorname{fourierbasis}_n(t) = a_n\operatorname{cos}(nt) + b_n\operatorname{sin}(nt)
$$


Next, the fourier basis is expressed in terms of unit direction vector $\vec{n}$ by using the following identities:
$$
\begin{align}
\operatorname{cos}(t)&=\vec{n}_x \\
\operatorname{sin}(t)&=\vec{n}_y
\end{align}
$$

We further use the [Mutiple-Angle Formulas](test.de):
$$
\begin{align}
\operatorname{cos}(nt) &= \sum_{k=0}^{\lfloor\frac{n}{2}\rfloor}(-1)^k\binom{n}{2k}\operatorname{sin}^{2k}(t)\operatorname{cos}^{n-2k}(t)\\
\operatorname{sin}(nt) &= \sum_{k=0}^{\lfloor\frac{n-1}{2}\rfloor}(-1)^k\binom{n}{2k+1}\operatorname{sin}^{2k+1}(t)\operatorname{cos}^{n-2k-1}(t)
\end{align}
$$

With these identities we can express the fourierbasis in terms of cartesian coordinates and read of the moment tensor components.

For the zero moment, we have:
$$
\operatorname{fourierbasis}_0(t) = a_0 \implies \mu_0[f] = a_0
$$

For the first moment, we get:
$$
\begin{align}
\operatorname{fourierbasis}_1(t) &= a_1\operatorname{cos}(t) + b_1\operatorname{sin}(t) \\
&= a_1\vec{n}_x + b_1\vec{n}_y \\
 &\implies \mu_1[f] = \left(\begin{matrix}
           a_{1} \\
           b_{1} \\
         \end{matrix}\right)
\end{align}
$$

For the second moment, we have:
$$
\begin{align}
\operatorname{fourierbasis}_2(t) &= a_2\operatorname{cos}(2t) + b_2\operatorname{sin}(2t) \\
&= a_2\left(\operatorname{cos}^2(t)-\operatorname{sin}^2(t)\right) + b_2\left(2\operatorname{cos}(t)\operatorname{sin}(t)\right) \\
&= a_2\vec{n}_x^2 - a_2\vec{n}_y^2 + b_2\vec{n}_x\vec{n}_y + b_2\vec{n}_y\vec{n}_x \\
 &\implies \mu_2[f] = \left(\begin{matrix}
           a_{2} & b_2 \\
           b_{2} & -a_2 \\
         \end{matrix}\right)
\end{align}
$$
