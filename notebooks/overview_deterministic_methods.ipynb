#### Transport equation (TE) forms
- **first order form** (classical radiative transfer equation)
    - streaming term of TE uses only first order spatial derivatives
    - works better for streaming dominated problems
- second order forms
    - streaming term of TE uses second order spatial derivatives
    - second order transport operators are symmetric positive definite (SPD)
    - works better for scattering dominated problems
    - **even-odd parity (EOP)**
        - does not work in voids, numerical issues near voids
        - parabolic second order PDE
        - presence of the total crosssection in the denominator of the streaming terms makes the treatment of voids difficult
    - **self-adjoint angular flux (SAAF)**
        - does not work in voids, numerical issues near voids
        - parabolic second order PDE
        - presence of the tota crosssection in the denominator of the streaming terms makes the treatment of voids difficult
    - **least squares/normal form (LS)**
        - not compatible with source iteration
        - applicable in voids since inverse of the removal/scattering operator is not needed
        - [least squares form compatible with voids](http://www.tandfonline.com/doi/abs/10.1080/00411450.2014.927364)
            - compatible with source iteration
            - the standard LS form has problems in highly diffuse media. This paper proposes a variation to the LS-form  which works well in both, highly diffuse media and voids


#### Discretizations in angular domain
- delta distributions with weight (discrete ordinates or SN)
    - disadvantages:
        - ray effect
- spherical harmonics (PN method and variations, depending on closure)
    - advantages:
        - rotationally invariant (avoids ray effect)
        - specific coupling structure between moments
    - disadvantages:
        - Gibbs phenomenon (in angular domain, but errors will be carried around in spatial domain)
- finite elements in angular domain














#### Notes:
- wether a method can deal with voids does not depend on the angular discretization, but rather the TE form used
- no single spatial discretization scheme that is appropriate for every transport problem
- transport equation can satisfy parabolic, elliptic and hyperbolic equations in certain physical limits
    - parabolic
    - elliptic
    - hyperbolic
- solution methods
    - source iteration
    - diffusion synthetic acceleration

### solution methods

- discontinuous finite element (a.k.a. discontinuous Galerkin method)
    - allows the field variable and/or its derivatives to be discontinous across element boundaries
    - combines features of finite element and finite volume methods
    - has finite volume and finite difference method as subset
