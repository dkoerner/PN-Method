<h1><center>Second-order forms of the transport equation</center></h1>

<h3><center>Even-odd parity (EOP) form of the transport equation</center></h3>

The EOP-form is a second order form of the transport equation, which means that the streaming term involves second-order spatial derivatives. This form is derived from the first order form of the transport equation by splitting the radiance field into even and odd components:

$$
\begin{align}
L(\vec{x}, \omega) &= L^-(\vec{x}, \omega) + L^+(\vec{x}, \omega)
\end{align}
$$

with

$$
\begin{align}
L^+(\vec{x}, \omega) &= L(\vec{x}, \omega) + L(\vec{x}, -\omega)\\
L^-(\vec{x}, \omega) &= L(\vec{x}, \omega) - L(\vec{x}, -\omega)
\end{align}
$$

Now we take the first order transport equation and replace $\omega$ by its negative counterpart $-\omega$:

$$
(-\omega\cdot\nabla)L(\vec{x}, -\omega)=-\sigma_t(\vec{x})L(\vec{x}, -\omega) + \sigma_s(\vec{x})\int_{\Omega'}f_p(\vec{x}, \omega'\cdot-\omega)L(\vec{x}, \omega')\mathbf{d}\omega' + Q(\vec{x})
$$

In the literature, scattering and emission is often assumed to be isotropic:

$$
(-\omega\cdot\nabla)L(\vec{x}, -\omega)=-\sigma_t(\vec{x})L(\vec{x}, -\omega) + \sigma_s(\vec{x})\phi(\vec{x}) + Q(\vec{x})
$$

We now produce the even-parity form of the transport equation by adding the equation above to the classical first order transport equation:

$$
\begin{align}
(\omega\cdot\nabla)L(\vec{x}, \omega) + (-\omega\cdot\nabla)L(\vec{x}, -\omega) + \sigma_t(\vec{x})L(\vec{x}, \omega) + \sigma_t(\vec{x})L(\vec{x}, -\omega) &= \sigma_s(\vec{x})\phi(\vec{x}) + \sigma_s(\vec{x})\phi(\vec{x}) + Q(\vec{x}) + Q(\vec{x})\\
(\omega\cdot\nabla)(\underbrace{L(\vec{x}, \omega) - L(\vec{x}, -\omega)}_{=L^-\left(\vec{x}, \omega\right)}) + \sigma_t(\vec{x})(\underbrace{L(\vec{x}, \omega) + L(\vec{x}, -\omega)}_{=L^+\left(\vec{x}, \omega\right)})&= 2\sigma_s(\vec{x})\phi(\vec{x}) + 2Q(\vec{x})\\
(\omega\cdot\nabla)L^-(\vec{x}, \omega) + \sigma_t(\vec{x})L^+\left(\vec{x}, \omega\right)&= 2\sigma_s(\vec{x})\phi(\vec{x}) + 2Q(\vec{x})
\end{align}
$$

Similarily, the odd-parity form of the transport equation is found be substraction:

$$
\begin{align}
(\omega\cdot\nabla)L(\vec{x}, \omega) - (-\omega\cdot\nabla)L(\vec{x}, -\omega) + \sigma_t(\vec{x})L(\vec{x}, \omega) - \sigma_t(\vec{x})L(\vec{x}, -\omega) &= \sigma_s(\vec{x})\phi(\vec{x}) - \sigma_s(\vec{x})\phi(\vec{x}) + Q(\vec{x}) - Q(\vec{x})\\
(\omega\cdot\nabla)(\underbrace{L(\vec{x}, \omega) + L(\vec{x}, -\omega)}_{=L^+\left(\vec{x}, \omega\right)}) + \sigma_t(\vec{x})(\underbrace{L(\vec{x}, \omega) - L(\vec{x}, -\omega)}_{=L^-\left(\vec{x}, \omega\right)})&= 0\\
(\omega\cdot\nabla)L^+(\vec{x}, \omega) + \sigma_t(\vec{x})L^-\left(\vec{x}, \omega\right)&= 0
\end{align}
$$

The odd-parity transport equation is resolved for $L^-$:

$$
L^-(\vec{x}, \omega) = -\frac{1}{\sigma_t\left(\vec{x}\right)}\left(\omega\cdot\nabla\right)L^+\left(\vec{x}, \omega\right)
$$

and substituted in the even-parity tranport equation to eliminate $L^-$. This produced the second-order even-odd transport equation, because the streaming term now contains second-order spatial derivatives:

$$
-\left(\omega\cdot\nabla\right)\left(\frac{1}{\sigma_t\left(\vec{x}\right)}\left(\omega\cdot\nabla\right)L^+\left(\vec{x}, \omega\right)\right)+ \sigma_t(\vec{x})L^+\left(\vec{x}, \omega\right)= 2\sigma_s(\vec{x})\phi(\vec{x}) + 2Q(\vec{x})
$$

- This equation is used to solve for $L^+$. With that, $L$ can be reconstructed using the odd-parity transport equation.
- Note that methods which are based on the EOP form of the transport equation will not work with vacuum regions, due to the $\sigma_t(\vec{x})$ term in the demoninator. 
- Also note that this is still completeley unrelated to the choice of discretization in spatial and angular domain.
- Methods based on the EOP-form need to be careful when specifying boundary conditions, as the unknown $L^+$ is only defined on a hemisphere.

<h3><center>Self-adjoint angular flux (SAAF) form of the transport equation</center></h3>

This form is found by rearranging the first order form (for isotopic source and scattering) to:

$$
L(\vec{x}, \omega) = \frac{\sigma_s(\vec{x})\phi(\vec{x}) + Q(\vec{x})}{4\pi\sigma_t(\vec{x})} - \frac{1}{\sigma_t(\vec{x})}(\omega\cdot\nabla)L(\vec{x}, \omega)
$$

Now this expression for $L$ is used with the first order transport equation. The $L$ in the streaming term is replaced by the expression above:

$$
(\omega\cdot\nabla)\left(\frac{\sigma_s(\vec{x})\phi(\vec{x}) + Q(\vec{x})}{4\pi\sigma_t(\vec{x})} - \frac{1}{\sigma_t(\vec{x})}(\omega\cdot\nabla)L(\vec{x}, \omega)\right)=-\sigma_t(\vec{x})L(\vec{x}, \omega) + \frac{\sigma_s(\vec{x})\phi(\vec{x})+Q(\vec{x})}{4\pi}
$$

which expands to the second-order self-adjoint angular flux form of the transport equation:

$$
 - (\omega\cdot\nabla)\frac{1}{\sigma_t(\vec{x})}(\omega\cdot\nabla)L(\vec{x}, \omega)=-\sigma_t(\vec{x})L(\vec{x}, \omega) + \frac{\sigma_s(\vec{x})\phi(\vec{x})+Q(\vec{x})}{4\pi}-(\omega\cdot\nabla)\frac{\sigma_s(\vec{x})\phi(\vec{x}) + Q(\vec{x})}{4\pi\sigma_t(\vec{x})}
$$

- Note how the second-order streaming term is similar to the streaming term in the EOP form.
- The SAAF form also will have problems with vacuum regions, due to the $\sigma_t$ term in the denominator.




<h3><center>Least-squares/normal form (LS) of the transport equation</center></h3>

The first order transport form is expressed in operator notation:

$$
TL + CL = SL + Q
$$

where $T$ is the transport or streaming operator which contains the directional derivative, $C$ is the collision or removal operator ($\sigma_t(\vec{x})$), $S$ is the scattering operator and $Q$ represents sources and emission.

The TE can be rearranged, such that the operators are collapsed into one:

$$
(T + C - S)L = Q
$$

In order to get a second-order form, one can multiply this equation by the adjoint of $T+C-S$ (which we denote $(T+C-S)^\star$). This gives the least-squares or normal form of the transport equation:

$$
(T + C - S)^\star(T + C - S)L = (T + C - S)^\star Q
$$

The structure is similar to the normal form $A^T A x = A^T b$ which is the reason for the name of this form.

It is easily verified that the transport term now becomes second-order:

$$
-(\omega\cdot\nabla)(\omega\cdot\nabla L)
$$

This term does not have the inverse of the extinction coefficient like EOP or SAAF and therefore this form can deal with vacuum regions.

Now the literature on LSPN is not consistent. They differ in the used form of the transport equation and the weight function. Here we show two different papers explaining their version of LSPN:

####  [A least-squares transport equation compatible with voids](The Least-Squares Finite Element Method):

First the paper rearranges the TE, such that the scattering operator remains on the RHS:

$$
(T + C )L = SL + Q
$$

In order to get a second-form, the paper multiplies this equation by the adjoint of $T+C$ (which we denote $(T+C)^\star$):

$$
(T + C )^\star(T + C)L = (T + C )^\star S L + (T + C )^\star Q
$$

Next, the paper multiplies with the basis function used for discretization.


This form is now used to derive a system of linear equations using the weighted-residual/finite element approach:

1. a spatial discretization is chosen 
* the angular domain is discretized into coefficients of some spherical basis functions $\psi_n$, which could be discrete ordinates ($S_N$), truncated spherical harmonics ($P_N$) or some other finite element discretization of the angular domain.
- Then weighting functions $w_n$ are found by applying the transport operator $(T+C-S)$ to all basis functions $\psi_n$:
$$
w_n(\vec{x}, \omega) = (T+C-S)\psi_n(\vec{x}, \omega)
$$
- Finally, the normal form of the transport equation is multiplied by these weighting functions and integrated over spatial and angular domain (on both sides). This produces a linear system of equations $Ax=b$, where A is of SPD (symmetric, positive, definit) form because the normal form of the transport equation is second order.


Notes:
- the scattering operator in the normal form of the transport equation causes source iteration to not work (probably because of negative main diagonal elements). An alternative normal form $(T+C)^\star(T+C-S)L = (T+C)^\star Q$ (?) which allows source iteration to be applied while still producing a SPD coefficient matrix. Therefore this approach may be suitable for a multigrid approach.
