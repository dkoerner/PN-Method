#### Notes:
- addressed some general questions which were expressed during the email conversation prior to the meeting
    - what do one looses when using SPN in comparison to PN?
        - SPN does not converge when the number of moments go to infinity (there is some bias)
            - however, the limit is reached very quickly after a short number of iterations
            - Martin showed me a PhD thesis which did a study on the error
                - reference needed p. 80
                - the plot showed that SPN error would drop quickly, but approach some error > 0 asymptotically
        - Martin also mentioned that, very roughly speaking, one could say that SPN is basically throwing away some coefficients in PN
        - it was also confirmed, that SPN can give results which have an approximation error which is worse than PN
        - but in some scenarios, SPN is giving the identical result to PN, such as:
            - an isotropic, inhomogeneous source
            - others, which weren't discussed
        - it was unclear, weather SPN works well in vacuum regions
            - in one of the emails, it was mentioned that SPN potentially works well with vaccum
            - in the literature, one finds that SPN is said to only work well in the diffusive regime (similar to diffusion)
                - for example _"Although there is a theoretical foundation for the SPn equations,
these equations are not effective for solving all types of
transport problems. As the derivations below will detail in a more
precise fashion, if the problem is not (1) “close” to diffusive or
(2) locally 1-D, SPn can give answers worse than diffusion"_ ([from here](http://www.tandfonline.com/doi/abs/10.1080/00411450.2010.535088?journalCode=ltty20))
        - we concluded that this is something which is best found by doing some studies with vacuum regions 


- current ideas/challenges for the use of starmap in computer graphics context:
    - Support for anisotropy
        - Martin explained, that the use of exponentials was motivated by the fact that the asymptotic preserving property (AP-property) should be retained. 
        - Other methods can also be used for the time integration, such as an explicit euler. In that case however, the AP property may not be retained and a limit on spatial- and time steps may be introduced.
        - The problem that C needs to be stored per voxel is also present in starmap.
    - Steady state
        - Martin mentioned, that the differential matrix equation could be solved if we just neglect the time derivative term (one can build a solution method for that). He briefly described how such a method could solve for even coefficients. These then would be used to find the values for odd coefficients. 
            - Kerstin noted, that the boundary condition are very important and for steady state solves will likely become unstable. If we use zero-value BC without any light being emitted from the boundary into the domain, it may work.
    - Support for vaccuum regions
        - we discussed that SPN and maybe other closures should tested with vacuum regions in order to get a better understanding about how well the different closures work with vacuum
        - other methods are much better suited for problems containing vacuum, such as least-squares PN or minimum entropy methods
        
        
- general exchange on current and alternative methods/approaches in neutron transort and light transport in computer graphics
    - least-squares PN
        - Martin mentioned that he has investigated or wants to investigate this method for his research
        - he briefly mentioned something related to a matrix, which needs to have a non-trivial kernel. This is violated in vacuum with pure transport and least-squares PN addresses this somehow (his explanation was good but I didn't get down the details)
        - the problem with least-squares PN is, that it is not energy-conserving and therefore approximative, which is a big deal for neutron transport. This is why it is not used much. Maybe it can be applied in graphics.
    - source iteration
        - recently I learned that lambda iteration is a method, which is very popular in astrophysics codes these days for radiative transfer computation (see chapter 4.4 in [this document](www.ita.uni-heidelberg.de/~dullemond/lectures/radtrans_2012/Chapter_4.pdf)). I asked if this method is also used or known in neutron transport. It may be the case that it is known under source iteration in neutron transport.
    - Martin asked about the importance of non-classical transport in graphics. I told him that it is not very popular, but I felt I couldn't give him a satisfactory answer.
    - Martin mentioned, that currently high-order/low-order methods (HiLo methods) are a hot topic in neutron transport.
        - the rough idea is to describe transport with some PDE (such as diffusion), but use a different method for computing the flux between neighbouring grid cells/voxels (e.g. use Monte-Carlo to compute the gradient from diffusion equation)
            - low order: diffusion; high order: monte carlo
    - I asked, how popular lattice boltzman methods are in neutron transport. Martin mentioned, that those were not very popular, since they always contain a diffusive term and therefore are better applied to problems with viscosity (such as fluids)
    - I asked about multi-grid methods in general for transport problems in neutron transport
         - Martin mentioned a paper, which showed that in order for multi-grid methods to work, the operator needs to contain some diffusion
         - since our proplem potentially has no diffusion, applying multi-grid in general could be tricky
     - on the other hand, Martin mentioned that people will certainly have tried multi-grid with PN and that there should be some related work
    - Gorji mentioned, that [this paper](http://dl.acm.org/citation.cfm?id=2792355) combines moment methods with variance reduction techniques
