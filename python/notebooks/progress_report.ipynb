# First order RTE form (steady state)

\begin{align}
(\omega\cdot\nabla)L(\vec{x}, \omega)=-\sigma_t(\vec{x})L(\vec{x}, \omega) + \sigma_s(\vec{x})\int_{\Omega'}f_p(\vec{x}, \omega'\cdot\omega)L(\vec{x}, \omega')\mathbf{d}\omega' + Q(\vec{x})
\end{align}

comlex-valued SH expansion in matrix form (where $\vec{L}$ and $\vec{u}$ are complex- and real-valued solution vectors respectively):
$$
M_0\partial_x\vec{L}+M_1\partial_y\vec{L} + M_2\vec{L} = \vec{q}
$$

real-valued sh expansion (as done in starmap):
$$
\left(S M_0S^{-1}\right)\partial_x\vec{u}+\left(S M_1S^{-1}\right)\partial_y\vec{u} + \left(S M_2S^{-1}\right)\vec{u} = S\vec{q}
$$

# Least-squares/second order RTE form (steady state)

$$
-\omega\cdot\nabla\left(\omega\cdot\nabla L\right)
-\underbrace{\left(\omega \cdot\nabla\sigma_t\right)L}_{\text{term 1}}
+\sigma_t^2 L
+\underbrace{\omega\cdot\nabla\left(\sigma_s\int_{\Omega'}f_p\left(\omega'\rightarrow\omega\right)L\mathbf{d}\omega\right)}_{\text{term 3}}
-\sigma_t\sigma_s\int_{\Omega'}{f_p\left(\omega'\rightarrow\omega\right)L\mathbf{d}\omega}
=
-\omega\cdot\nabla Q
+\sigma_t Q
$$


comlex-valued SH expansion in matrix form (where $\vec{L}$ and $\vec{u}$ are complex- and real-valued solution vectors respectively):
$$
M_0\partial_x\partial_x\vec{L}+
M_1\partial_x\partial_y\vec{L}+
M_2\partial_y\partial_x\vec{L}+
M_3\partial_y\partial_y\vec{L}+
M_4\partial_x\vec{L}+
M_5\partial_y\vec{L}+
M_6\vec{L}
= \vec{q}
$$

real-valued sh expansion:
$$
\left(S M_0S^{-1}\right)\partial_x\partial_x\vec{u}+
\left(S M_1S^{-1}\right)\partial_x\partial_y\vec{u}+
\left(S M_2S^{-1}\right)\partial_y\partial_x\vec{u}+
\left(S M_3S^{-1}\right)\partial_y\partial_y\vec{u}+
\left(S M_4S^{-1}\right)\partial_x\vec{u}+
\left(S M_5S^{-1}\right)\partial_y\vec{u}+
\left(S M_6S^{-1}\right)\vec{u}
= S \vec{q}
$$
